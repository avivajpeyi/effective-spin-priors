```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
from priors import *
```

```python
fig = plt.figure(figsize=(15,4))

ax = fig.add_subplot(131)

aMax=1
q=0.8
ntrials=100000
s1s = aMax*(2.*np.random.random(ntrials)-1.)
s2s = aMax*(2.*np.random.random(ntrials)-1.)
xeff = (s1s + q*s2s)/(1.+q)
xs = np.linspace(-1,1,300)
p_xeff = chi_effective_prior_from_aligned_spins(q,aMax,xs)

ax.hist(xeff,density=True,bins=50)
ax.plot(xs,p_xeff,color='black')
ax.xaxis.grid(True,which='major',ls=':',color='grey')
ax.yaxis.grid(True,which='major',ls=':',color='grey')
ax.tick_params(labelsize=14)
ax.set_xlabel(r'$\chi_{\rm eff}$',fontsize=18)
ax.set_ylabel(r'$p(\chi_{\rm eff})$',fontsize=18)
ax.annotate(r'$q={0}$'.format(q),(0.75,0.85),xycoords='axes fraction',fontsize=16)
ax.annotate(r'$a_{{\rm max}} = {0}$'.format(aMax),(0.75,0.75),xycoords='axes fraction',fontsize=16)

ax = fig.add_subplot(132)

aMax=1
q=0.1
ntrials=100000
s1s = aMax*(2.*np.random.random(ntrials)-1.)
s2s = aMax*(2.*np.random.random(ntrials)-1.)
xeff = (s1s + q*s2s)/(1.+q)
xs = np.linspace(-1,1,300)
p_xeff = chi_effective_prior_from_aligned_spins(q,aMax,xs)

ax.hist(xeff,density=True,bins=50)
ax.plot(xs,p_xeff,color='black')
ax.xaxis.grid(True,which='major',ls=':',color='grey')
ax.yaxis.grid(True,which='major',ls=':',color='grey')
ax.tick_params(labelsize=14)
ax.set_xlabel(r'$\chi_{\rm eff}$',fontsize=18)
ax.annotate(r'$q={0}$'.format(q),(0.75,0.85),xycoords='axes fraction',fontsize=16,backgroundcolor=(1,1,1,0.95))
ax.annotate(r'$a_{{\rm max}} = {0}$'.format(aMax),(0.75,0.75),xycoords='axes fraction',fontsize=16,backgroundcolor=(1,1,1,0.95))
ax.set_ylim(0,0.65)

ax = fig.add_subplot(133)

aMax=0.4
q=0.8
ntrials=30000
s1s = aMax*(2.*np.random.random(ntrials)-1.)
s2s = aMax*(2.*np.random.random(ntrials)-1.)
xeff = (s1s + q*s2s)/(1.+q)
xs = np.linspace(-1,1,300)
p_xeff = chi_effective_prior_from_aligned_spins(q,aMax,xs)

ax.hist(xeff,density=True,bins=50)
ax.plot(xs,p_xeff,color='black')
ax.xaxis.grid(True,which='major',ls=':',color='grey')
ax.yaxis.grid(True,which='major',ls=':',color='grey')
ax.tick_params(labelsize=14)
ax.set_xlabel(r'$\chi_{\rm eff}$',fontsize=18)
ax.annotate(r'$q={0}$'.format(q),(0.7,0.85),xycoords='axes fraction',fontsize=16)
ax.annotate(r'$a_{{\rm max}} = {0}$'.format(aMax),(0.7,0.75),xycoords='axes fraction',fontsize=16)

plt.tight_layout()
plt.savefig('demo_chi_eff_aligned.pdf',bbox_inches='tight')

plt.show()
```
